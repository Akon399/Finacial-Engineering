# BSM Dividend model.
# Cal d1 & d2.

```python
import numpy as np
from scipy.stats import norm

N = norm.cdf
def BS_CALL(S,K,T,r,q,sigma):
    d1 = (np.log(S/K) + ((r-q) + sigma**2/2)*T) / sigma * T**0.5
    d2 = d1 - sigma * np.int(T**0.5)
    return S * np.exp(-q*T) * N(d1) - K * np.exp(-r*T) * N(d2)
```

```python
# Now lets plot our calls d with MatplotLib.

import matplotlib.pyplot as plt

K = 100
r = 0.1
q = 0.05
T = 1
sigma = 0.3

S = np.arange(60,140,0.1) # Array.

calls = [BS_CALL(s,K,T,r, q,sigma) for s in S] # Array of Series of option value.
callsD = [BS_CALL_D(s,K,T,r,q,sigma) for s in S]

plt.plot(s,calls, label = 'Call value')
plt.plot(s,callsD,label = 'Put value')
plt.xlabel('Stock Price')
plt.ylabel('Option value')
plt.title('Impact of Blacl Scholes Model of S')
```

```python
# BSM model with dividend.
# Written notes for Lecture 45 by Akhona Njeje on 5 January 2022.

import numpy as np
from scipy.stats import norm

N = norm.cdf

def BS_CALL_D(S, K, T, r, q, sigma):
    d1 = (np.log(S/K) + ((r-q) + sigma**2/2)*T) / sigma * np.sqrt(T)  # Modelling.
    d2 = d1 - sigma * np.sqrt(T)
    return S * np.exp(-q*T)*N(d1) - K * np.exp(-r*T) * N(d2)
```
